```python
from typing import List, Optional
from pydantic import BaseModel, ValidationError

# Modelleri tanımla
class Bolge(BaseModel):
    id: int
    bolge: str

    class Config:
        orm_mode = True
        allow_population_by_field_name = True

class IsYeri(BaseModel):
    id: int
    sirketId: Optional[int] = None
    bolgeId: Optional[int] = None
    sapKodu: str
    isYeriAdi: Optional[str] = None
    kullaniciAdi: Optional[str] = None
    kullaniciKodu: Optional[str] = None
    sistemSifresi: Optional[str] = None
    isYeriSifresi: Optional[str] = None
    sgkSicilNo: Optional[str] = None
    isYeriAdresi: Optional[str] = None
    acilisTarihi: Optional[str] = None
    kapanisTarihi: Optional[str] = None
    bolgeler: list[Bolge]=[]

    class Config:
        orm_mode = True
        allow_population_by_field_name = True

class Proje(BaseModel):
    id: int
    isYeriId: Optional[int] = None
    pypOgesi: Optional[str] = None

    class Config:
        orm_mode = True
        allow_population_by_field_name = True

class Sirket(BaseModel):
    id: int
    sirketKodu: int
    sirketAdi: str

    class Config:
        orm_mode = True
        allow_population_by_field_name = True
```
